```python
#lets import the libraries
import pandas as pd
import numpy as np
```

```python
              # data-collection
#lets read the data-set by declaring data-frame (true and fake) for true and fake dataset
true = pd.read_csv('/content/True.csv')
fake = pd.read_csv('/content/Fake.csv')
```

```python
#lets create a cloumn in the true and fake dataset to provide that this content of data is true or false (1,0)
true['target'] = 1
fake['target'] = 0
```

```python
#view of data-set
true.head()
```

```python
fake.head()
```

```python
#lets conacatenate the data-set of true and fake
news = pd.concat([true,fake],axis=0)
```

```python
#lets view
news.head()    #true news data-set view
```

```python
news.tail()   #fake news data-set view
```

```python
#before lets delete the useless column not required for the data-sey
news = news.drop(['title','subject','date'],axis=1)
```

```python
#lets check the new form of data-set
news.head()
```

```python
# @title as we cana see the the true news is preset in the top most level of the data-set and fake news present at down level of the news data-set
```

```python
#                     data preprocessing
#lets reshuffle the data-set
news = news.sample(frac=1)
```

```python
news.head()
```

```python
news.reset_index(inplace=True) #cerating an index column
```

```python
news.head()
```

```python
#cleaning of data
#lets remove the index as its of no use
news.drop(['index'],axis=1, inplace = True)
```

```python
#lets check data-set
news.head()
```

```python
# @title our data is read
```

```python
#      data-future extraction
import re # covent text into readable form
```

```python
#lets defie the function
def wordopt(text):    #fuction = wordopt with respect to text column of the data-set
  text = text.lower()

  #removing the html tag
  text = re.sub('\[.*?\]', '', text)

  #remopving punctuation
  text = re.sub("\\W"," ",text)

  #removing the urls
  text = re.sub('https?://\S+|www\.\S+', '', text)

  #remove the didgits
  text = re.sub('\d', '', text)

  return text
```

```python
news['text'].apply(wordopt)   #here we are applying the wordopt fuction with respect to the text column og the news data set
```

```python
#lets declare the x and y parameters for training and  testing the data-set
x = news['text']
y = news['target']
```

```python
from sklearn.model_selection import train_test_split

#now lets do the opration
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)      #operation done with respect to size to 30%
```

```python
#lets check the shape
x_train.shape
```

```python
y_train.shape
```

```python
#lets create vectors for the ml model
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vector = TfidfVectorizer()
```

```python
#now lests work on the oprtaion of vector with respect x and y values
xv_train = vector.fit_transform(x_train)

#lest sort the data-set
xv_test = vector.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
# now after performing the vectors for the x and y variables now lets
#lets import the ml model
```

```python
#logistic regression
import sklearn.linear_model as LogisticRegression
LR = LogisticRegression.LogisticRegression()
```

```python
#lets fit for the vectors xv_trian and xv_test
LR.fit(xv_train,y_train)
```

```python
#lets predict the value for the logistic regression XV_test vector
LR.predict(xv_test)
```

```python
#lets see the score
LR.score(xv_test,y_test)              #value is recoginze upto to 98%  preety ggod score for the model
```

```python
# @title decision tree clasisfier
#decision tree classifier
```

```python
import sklearn.tree as DecisionTreeClassifier
DT = DecisionTreeClassifier.DecisionTreeClassifier()
```

```python
#LETS fit the vector into th DT model
DT.fit(xv_train,y_train)
```

```python
#predict the DTC as array or not
DT.predict(xv_test)
```

```python
#predic the score for DT model
DT.score(xv_test,y_test)         # value is upto 99% preety a good score
```

```python
# @title Random forest tree
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()

#lets fit the vecors (X,Y) to the RF
RF.fit(xv_train,y_train)                 #take tooo much time for fitting the data-set in the randomforestclassifier
```

```python
#lets predict the RF
RF.predict(xv_test)
```

```python
#lets see the score for the model
RF.score(xv_test,y_test)         #98% successful
```

```python
# lets check the prediction model
def output_lable(n):
  if n == 0:
    return "Fake News"
  elif n == 1:
    return "Not A Fake News"
```

```python
#lets declare the another function for manual testing of the data
def manual_testing(news):
    testing_news = {"text": [news]} #corrected sytax for defining dictionary
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vector.transform(new_x_test)
    #prdicted values
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_RF = RF.predict(new_xv_test)

    return ("\n\nLR Prediction: {} \nDT Prediction: {} \nRF Prediction: {}".format(output_lable(pred_LR[0]), output_lable(pred_DT[0]), output_lable(pred_RF[0])))
```
